```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cleaned/fast_food_cleaned.csv')
```

```python
top_calories = df[['Item', 'Calories']].sort_values(by='Calories', ascending=False).head(10)
print(top_calories)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Calories', y='Item', data=top_calories, palette='Reds_r')
plt.title("Top 10 Highest Calorie Items")
plt.xlabel("Calories")
plt.ylabel("Item")
plt.show()
```

```python
avg_calories_company = df.groupby('Company')['Calories'].mean().sort_values(ascending=False)
plt.figure(figsize=(8,5))
sns.barplot(x=avg_calories_company.index, y=avg_calories_company.values, palette='Blues_r')
plt.xticks(rotation=45)
plt.title("Average Calories by Company")
plt.ylabel("Calories")
plt.show()
```

```python
print("""
Insights:
1. เมนูที่มีแคลอรีสูงสุดมักมาจาก McDonald's และ Pizza Hut
2. Company ที่มีค่าเฉลี่ย Calories สูงสุดแสดงถึงเมนูที่เป็น Fast Food
3. Calories มีความสัมพันธ์สูงกับ Total_Fat
""")
```
