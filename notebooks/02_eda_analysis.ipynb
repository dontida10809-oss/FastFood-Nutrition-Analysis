```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cleaned/fast_food_cleaned.csv')
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Calories'], bins=20, kde=True)
plt.title("Calories Distribution")
plt.xlabel("Calories")
plt.ylabel("Count")
plt.show()
```

```python
avg_calories = df.groupby('Company')['Calories'].mean().sort_values(ascending=False)
print(avg_calories)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=avg_calories.index, y=avg_calories.values)
plt.xticks(rotation=45)
plt.title("Average Calories by Company")
plt.ylabel("Average Calories")
plt.show()
```
