```python
import pandas as pd

# โหลดไฟล์ CSV
df = pd.read_csv('../data/raw/fast_food_nutrition.csv')

# เปลี่ยนชื่อคอลัมน์ให้เรียบง่ายและมาตรฐาน
df.rename(columns={
    'Calories from\nFat': 'Calories_from_Fat',
    'Total Fat\n(g)': 'Total_Fat',
    'Saturated Fat\n(g)': 'Saturated_Fat',
    'Trans Fat\n(g)': 'Trans_Fat',
    'Cholesterol\n(mg)': 'Cholesterol',
    'Sodium \n(mg)': 'Sodium',
    'Carbs\n(g)': 'Carbohydrates',
    'Fiber\n(g)': 'Fiber',
    'Sugars\n(g)': 'Sugars',
    'Protein\n(g)': 'Protein',
    'Weight Watchers\nPnts': 'Weight_Watchers_Points'
}, inplace=True)

# ตรวจสอบคอลัมน์ใหม่
print(df.columns.tolist())
```

```python
numeric_cols = ['Calories', 'Calories_from_Fat', 'Total_Fat', 'Saturated_Fat', 'Trans_Fat',
                'Cholesterol', 'Sodium', 'Carbohydrates', 'Fiber', 'Sugars', 'Protein']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# ลบค่า NaN ที่เกิดจากการ convert
df.dropna(subset=numeric_cols, inplace=True)

# บันทึกไฟล์ที่ทำความสะอาดแล้ว
df.to_csv('../data/cleaned/fast_food_cleaned.csv', index=False)
print("Data cleaning complete. Saved to cleaned folder.")
```
